```python
# Lendo a base de dados

import pandas as pd
obesity = pd.read_csv(r"https://github.com/OlicerioN/tech_challenge_4/blob/main/Obesity.csv", sep = ',')
obesity.head()
```

```python
# Verificando tipagem

obesity.info()
```

```python
# Verificando características de colunas numéricas

obesity.describe()
```

```python
# Verificando nulos por coluna

nulls = obesity.isnull().sum()
print(nulls)
```

```python
# Verificando linhas duplicadas

print("Dup 1:", obesity.duplicated().sum())

# Removendo linhas duplicadas

obesity = obesity.drop_duplicates()

# Confirmando remoção das linhas duplicadas

print("Dup 2:", obesity.duplicated().sum())
```

```python
# Verificando valores possíveis e quantidades da variável Target

obesity['Obesity'].value_counts()
```

```python
## Arredondar colunas numéricas decimais para número inteiro, conforme indicado no dicionário disponibilizado

# 1 - selecionando colunas numéricas

col_Int_double = ['FCVC','NCP','CH2O','FAF','TUE']

obesity[col_Int_double] = obesity[col_Int_double].round().astype(int)

# 3 - verificando resultado

obesity.head()
```

```python
# Usamos como boas práticas transformar variáveis "Sim" e "Não" em variáveis booleanas

obesity['Gender'] = obesity['Gender'].replace({'Female': 0, 'Male': 1}).astype(int)
obesity['family_history'] = obesity['family_history'].replace({'no': 0, 'yes': 1}).astype(int)
obesity['FAVC'] = obesity['FAVC'].replace({'no': 0, 'yes': 1}).astype(int)
obesity['SMOKE'] = obesity['SMOKE'].replace({'no': 0, 'yes': 1}).astype(int)
obesity['SCC'] = obesity['SCC'].replace({'no': 0, 'yes': 1}).astype(int)

# Verificando resultado

obesity.head()
```

```python
# Traduzindo cabeçalhos para português (BR)

obesity = obesity.rename(columns={'Gender': 'Gênero'})
obesity = obesity.rename(columns={'Age': 'Idade'})
obesity = obesity.rename(columns={'Height': 'Altura (m)'})
obesity = obesity.rename(columns={'Weight': 'Peso (kg)'})
obesity = obesity.rename(columns={'family_history': 'Histórico familiar'})
obesity = obesity.rename(columns={'FAVC': 'Frequência de consumo de alimentos calóricos'})
obesity = obesity.rename(columns={'FCVC': 'Frequência de consumo de vegetais'})
obesity = obesity.rename(columns={'NCP': 'Número de refeições diárias'})
obesity = obesity.rename(columns={'CAEC': 'Consumo de lanches entre refeições'})
obesity = obesity.rename(columns={'SMOKE': 'Fumante'})
obesity = obesity.rename(columns={'CH2O': 'Consumo diário de água'})
obesity = obesity.rename(columns={'SCC': 'Monitora ingestão calórica diária'})
obesity = obesity.rename(columns={'FAF': 'Frequência semanal de exercícios'})
obesity = obesity.rename(columns={'TUE': 'Tempo diário usando dispositivos eletrônicos'})
obesity = obesity.rename(columns={'CALC': 'Consumo de bebida alcoólica'})
obesity = obesity.rename(columns={'MTRANS': 'Meio de transporte habitual'})
obesity = obesity.rename(columns={'Obesity': 'Nível de obesidade'})

# Verificando resultado

obesity.head()
```

```python
# Traduzindo valores em inglês das colunas qualitativas com base no dicionário

obesity['Consumo de lanches entre refeições'] = obesity['Consumo de lanches entre refeições'].replace({'Sometimes': 'pouco', 'Frequently': 'frequente', 'Always': 'sempre', 'no': 'nunca'})
obesity['Consumo de bebida alcoólica'] = obesity['Consumo de bebida alcoólica'].replace({'Sometimes': 'pouco', 'Frequently': 'frequente', 'Always': 'sempre', 'no': 'nunca'})
obesity['Meio de transporte habitual'] = obesity['Meio de transporte habitual'].replace({'Automobile': 'carro', 'Motorbike': 'moto', 'Bike': 'bicicleta', 'Public_Transportation': 'transporte público', 'Walking': 'caminhando'})
obesity['Nível de obesidade'] = obesity['Nível de obesidade'].replace({'Normal_Weight': 'peso normal', 'Overweight_Level_I': 'sobrepeso nível 1', 'Overweight_Level_II': 'sobrepeso nível 2', 'Obesity_Type_I': 'obesidade tipo 1', 'Obesity_Type_II': 'obesidade tipo 2', 'Obesity_Type_III': 'obesidade tipo 3', 'Insufficient_Weight': 'abaixo do peso'})

obesity.head()
```
